```python
# balance_optimizer.py
import openpyxl
import numpy as np
from pymoo.algorithms.moo.nsga2 import NSGA2
from pymoo.optimize import minimize
from pymoo.core.problem import Problem
import os
import shutil
from datetime import datetime

class GalacticSiegeBalancer:
    """
    Optimizador de balance para Galactic Siege usando NSGA-II
    Basado en el paper: Improved Game Units Balancing Through Combinatorial Optimization
    """
    
    def __init__(self, excel_path='game_balance.xlsx'):
        self.excel_path = excel_path
        self.backup_path = f"backup_{datetime.now().strftime('%Y%m%d_%H%M%S')}.xlsx"
        
        # Crear backup
        if os.path.exists(excel_path):
            shutil.copy(excel_path, self.backup_path)
            print(f"âœ… Backup creado: {self.backup_path}")
        
        self.wb = None
        self.config_sheet = None
        self.multilevel_sheet = None
        
    def load_workbook(self):
        """Carga el workbook de Excel"""
        self.wb = openpyxl.load_workbook(self.excel_path, data_only=False)
        self.config_sheet = self.wb['Config']
        self.multilevel_sheet = self.wb['MultiLevel']
        
    def save_workbook(self):
        """Guarda el workbook"""
        self.wb.save(self.excel_path)
        
    def read_current_config(self):
        """Lee la configuraciÃ³n actual del Excel"""
        config = {
            'player_health_l12': self.config_sheet['B2'].value,
            'player_health_l3': self.config_sheet['B3'].value,
            'player_bullet_damage': self.config_sheet['B4'].value,
            'player_fire_rate': self.config_sheet['B5'].value,
            'enemy1_health': self.config_sheet['B10'].value,
            'enemy1_spawn_interval': self.config_sheet['B14'].value,
            'enemy1_fire_rate': self.config_sheet['B13'].value,
            'enemy2_health': self.config_sheet['B19'].value,
            'enemy2_spawn_interval': self.config_sheet['B23'].value,
            'enemy3_health': self.config_sheet['B28'].value,
            'enemy3_spawn_interval': self.config_sheet['B32'].value,
            'enemy3_fire_rate': self.config_sheet['B31'].value,
            'boss_health': self.config_sheet['B37'].value,
            'extra_life_interval': self.config_sheet['B47'].value,
            'nuke_interval': self.config_sheet['B48'].value,
        }
        return config
    
    def write_config(self, config_array):
        """
        Escribe una configuraciÃ³n al Excel
        config_array: numpy array con 15 valores
        """
        self.load_workbook()
        
        # Mapeo de Ã­ndices a celdas del Config sheet
        mapping = [
            ('B2', int(config_array[0])),   # player_health_l12
            ('B3', int(config_array[1])),   # player_health_l3
            ('B4', int(config_array[2])),   # player_bullet_damage
            ('B5', round(config_array[3], 1)),  # player_fire_rate
            ('B10', int(config_array[4])),  # enemy1_health
            ('B14', int(config_array[5])),  # enemy1_spawn_interval
            ('B13', round(config_array[6], 1)),  # enemy1_fire_rate
            ('B19', int(config_array[7])),  # enemy2_health
            ('B23', int(config_array[8])),  # enemy2_spawn_interval
            ('B28', int(config_array[9])),  # enemy3_health
            ('B32', int(config_array[10])), # enemy3_spawn_interval
            ('B31', round(config_array[11], 1)),  # enemy3_fire_rate
            ('B37', int(config_array[12])), # boss_health
            ('B47', int(config_array[13])), # extra_life_interval
            ('B48', int(config_array[14])), # nuke_interval
        ]
        
        for cell, value in mapping:
            self.config_sheet[cell] = value
        
        self.save_workbook()
        
        # Recargar con data_only=True para obtener valores calculados
        self.wb = openpyxl.load_workbook(self.excel_path, data_only=True)
        self.multilevel_sheet = self.wb['MultiLevel']
    
    def read_objectives(self):
        """
        Lee los 4 objetivos calculados del Excel
        Returns: [obj1, obj2, obj3, obj4]
        """
        try:
            obj1 = self.multilevel_sheet['B24'].value  # SD Difficulty
            obj2 = self.multilevel_sheet['B25'].value  # Survival Gap
            obj3 = self.multilevel_sheet['B26'].value  # SD TTK
            obj4 = self.multilevel_sheet['B27'].value  # Score Deviation
            
            # Manejar valores None o errores
            objectives = []
            for obj in [obj1, obj2, obj3, obj4]:
                if obj is None or isinstance(obj, str):
                    objectives.append(999.0)  # PenalizaciÃ³n alta
                else:
                    objectives.append(float(obj))
            
            return objectives
        except Exception as e:
            print(f"âŒ Error leyendo objetivos: {e}")
            return [999.0, 999.0, 999.0, 999.0]
    
    def evaluate_config(self, config_array):
        """EvalÃºa una configuraciÃ³n y retorna los objetivos"""
        self.write_config(config_array)
        objectives = self.read_objectives()
        return objectives
    
    def print_config(self, config_array, objectives):
        """Imprime una configuraciÃ³n de forma legible"""
        print("\n" + "="*60)
        print("CONFIGURACIÃ“N:")
        print(f"  Player Health (L1-L2): {int(config_array[0])}")
        print(f"  Player Health (L3): {int(config_array[1])}")
        print(f"  Player Bullet Damage: {int(config_array[2])}")
        print(f"  Player Fire Rate: {config_array[3]:.1f} sec")
        print(f"  Enemy1 Health: {int(config_array[4])}")
        print(f"  Enemy1 Spawn Interval: {int(config_array[5])} sec")
        print(f"  Enemy1 Fire Rate: {config_array[6]:.1f} sec")
        print(f"  Enemy2 Health: {int(config_array[7])}")
        print(f"  Enemy2 Spawn Interval: {int(config_array[8])} sec")
        print(f"  Enemy3 Health: {int(config_array[9])}")
        print(f"  Enemy3 Spawn Interval: {int(config_array[10])} sec")
        print(f"  Enemy3 Fire Rate: {config_array[11]:.1f} sec")
        print(f"  Boss Health: {int(config_array[12])}")
        print(f"  Extra Life Interval: {int(config_array[13])} sec")
        print(f"  Nuke Interval: {int(config_array[14])} sec")
        
        print("\nOBJETIVOS:")
        print(f"  SD Difficulty: {objectives[0]:.3f} (target: < 0.10)")
        print(f"  Survival Gap: {objectives[1]:.3f} (target: < 5)")
        print(f"  SD TTK: {objectives[2]:.3f} (target: < 0.30)")
        print(f"  Score Deviation: {objectives[3]:.3f} (target: < 0.20)")
        print("="*60)


class BalanceOptimizationProblem(Problem):
    """Problema de optimizaciÃ³n multi-objetivo para NSGA-II"""
    
    def __init__(self, balancer):
        self.balancer = balancer
        
        # Definir lÃ­mites de variables (xl = lower, xu = upper)
        super().__init__(
            n_var=15,  # 15 variables de configuraciÃ³n
            n_obj=4,   # 4 objetivos a minimizar
            n_constr=1,  # 1 restricciÃ³n (survival > 0)
            xl=np.array([
                3,    # player_health_l12 min
                5,    # player_health_l3 min
                1,    # player_bullet_damage min
                0.3,  # player_fire_rate min
                1,    # enemy1_health min
                3,    # enemy1_spawn_interval min
                1.0,  # enemy1_fire_rate min
                1,    # enemy2_health min
                3,    # enemy2_spawn_interval min
                3,    # enemy3_health min
                8,    # enemy3_spawn_interval min
                2,    # enemy3_fire_rate min
                30,   # boss_health min
                10,   # extra_life_interval min
                30,   # nuke_interval min
            ]),
            xu=np.array([
                10,   # player_health_l12 max
                15,   # player_health_l3 max
                3,    # player_bullet_damage max
                1.0,  # player_fire_rate max
                6,    # enemy1_health max
                8,    # enemy1_spawn_interval max
                3.0,  # enemy1_fire_rate max
                5,    # enemy2_health max
                8,    # enemy2_spawn_interval max
                10,   # enemy3_health max
                16,   # enemy3_spawn_interval max
                6,    # enemy3_fire_rate max
                60,   # boss_health max
                30,   # extra_life_interval max
                60,   # nuke_interval max
            ])
        )
        
        self.evaluation_count = 0
        
    def _evaluate(self, X, out, *args, **kwargs):
        """
        EvalÃºa un conjunto de soluciones
        X: matriz donde cada fila es una configuraciÃ³n
        """
        objectives_list = []
        constraints_list = []
        
        for i, config in enumerate(X):
            self.evaluation_count += 1
            print(f"\nğŸ“Š Evaluando configuraciÃ³n {self.evaluation_count}/{len(X)}...")
            
            # Evaluar configuraciÃ³n
            objectives = self.balancer.evaluate_config(config)
            objectives_list.append(objectives)
            
            # RestricciÃ³n: Survival Gap debe ser positivo o pequeÃ±o negativo
            # Si es muy negativo = player muere demasiado rÃ¡pido
            constraint = max(0, objectives[1] - 60)  # Penalizar si gap > 60
            constraints_list.append([constraint])
            
            # Imprimir progreso
            if self.evaluation_count % 10 == 0:
                print(f"âœ… Evaluadas {self.evaluation_count} configuraciones")
        
        out["F"] = np.array(objectives_list)
        out["G"] = np.array(constraints_list)


def run_optimization():
    """FunciÃ³n principal para ejecutar la optimizaciÃ³n"""
    
    print("ğŸš€ GALACTIC SIEGE - BALANCE OPTIMIZER")
    print("Basado en: Improved Game Units Balancing Through Combinatorial Optimization")
    print("="*70)
    
    # Inicializar balanceador
    balancer = GalacticSiegeBalancer('game_balance.xlsx')
    
    # Leer configuraciÃ³n actual
    print("\nğŸ“– Leyendo configuraciÃ³n actual...")
    balancer.load_workbook()
    current_config = balancer.read_current_config()
    print("âœ… ConfiguraciÃ³n actual cargada")
    
    # Crear problema de optimizaciÃ³n
    problem = BalanceOptimizationProblem(balancer)
    
    # Configurar algoritmo NSGA-II (como en el paper)
    print("\nâš™ï¸ Configurando NSGA-II...")
    print("   PoblaciÃ³n: 50")
    print("   Generaciones: 20")
    print("   Total evaluaciones: 1000")
    
    algorithm = NSGA2(
        pop_size=50,
        eliminate_duplicates=True
    )
    
    # Ejecutar optimizaciÃ³n
    print("\nğŸ”„ Iniciando optimizaciÃ³n...")
    print("   Esto puede tomar 10-20 minutos dependiendo de tu hardware")
    print("   (El paper usÃ³ 2310 evaluaciones en 13 minutos)")
    
    try:
        result = minimize(
            problem,
            algorithm,
            ('n_gen', 20),
            seed=42,
            verbose=True
        )
        
        print("\nâœ… OPTIMIZACIÃ“N COMPLETADA")
        print("="*70)
        
        # Analizar resultados
        analyze_results(result, balancer)
        
    except Exception as e:
        print(f"\nâŒ Error durante optimizaciÃ³n: {e}")
        import traceback
        traceback.print_exc()


def analyze_results(result, balancer):
    """Analiza y presenta los resultados de la optimizaciÃ³n"""
    
    print("\nğŸ“Š ANÃLISIS DE RESULTADOS")
    print("="*70)
    
    # Obtener Pareto Front
    pareto_front = result.F
    pareto_solutions = result.X
    
    print(f"\nâœ… Soluciones en Pareto Front: {len(pareto_front)}")
    
    # Encontrar mejor soluciÃ³n (mÃ¡s cercana a origen)
    distances = np.sqrt(np.sum(pareto_front**2, axis=1))
    best_idx = np.argmin(distances)
    
    best_solution = pareto_solutions[best_idx]
    best_objectives = pareto_front[best_idx]
    
    print("\nğŸ† MEJOR SOLUCIÃ“N (mÃ¡s cercana a origen):")
    balancer.print_config(best_solution, best_objectives)
    
    # Calcular mejora vs configuraciÃ³n inicial
    print("\nğŸ“ˆ COMPARACIÃ“N CON CONFIGURACIÃ“N INICIAL:")
    
    # Leer objetivos de configuraciÃ³n inicial
    initial_config_array = np.array([
        5, 10, 1, 0.5,  # player
        3, 5, 1.5,      # enemy1
        2, 5,           # enemy2
        6, 12, 4,       # enemy3
        40, 15, 45      # boss, powerups
    ])
    
    initial_objectives = balancer.evaluate_config(initial_config_array)
    
    print("\nInicial:")
    print(f"  SD Difficulty: {initial_objectives[0]:.3f}")
    print(f"  Survival Gap: {initial_objectives[1]:.3f}")
    print(f"  SD TTK: {initial_objectives[2]:.3f}")
    print(f"  Score Deviation: {initial_objectives[3]:.3f}")
    
    print("\nOptimizada:")
    print(f"  SD Difficulty: {best_objectives[0]:.3f} ({improvement_pct(initial_objectives[0], best_objectives[0])})")
    print(f"  Survival Gap: {best_objectives[1]:.3f} ({improvement_pct(initial_objectives[1], best_objectives[1])})")
    print(f"  SD TTK: {best_objectives[2]:.3f} ({improvement_pct(initial_objectives[2], best_objectives[2])})")
    print(f"  Score Deviation: {best_objectives[3]:.3f} ({improvement_pct(initial_objectives[3], best_objectives[3])})")
    
    # Mostrar top 5 soluciones
    print("\nğŸ¯ TOP 5 SOLUCIONES DEL PARETO FRONT:")
    top_5_indices = np.argsort(distances)[:5]
    
    for rank, idx in enumerate(top_5_indices, 1):
        print(f"\n--- SoluciÃ³n #{rank} (distancia: {distances[idx]:.3f}) ---")
        config = pareto_solutions[idx]
        objectives = pareto_front[idx]
        print(f"SD Diff: {objectives[0]:.3f} | Survival: {objectives[1]:.3f} | SD TTK: {objectives[2]:.3f} | Score: {objectives[3]:.3f}")
    
    # Guardar mejores soluciones en Excel
    save_solutions_to_excel(balancer, pareto_solutions, pareto_front, best_idx)
    
    # Generar grÃ¡ficos
    plot_pareto_front(pareto_front, best_idx)


def improvement_pct(initial, optimized):
    """Calcula el porcentaje de mejora"""
    if initial == 0:
        return "N/A"
    improvement = ((initial - optimized) / initial) * 100
    if improvement > 0:
        return f"â†“ {improvement:.1f}% mejor"
    else:
        return f"â†‘ {abs(improvement):.1f}% peor"


def save_solutions_to_excel(balancer, solutions, objectives, best_idx):
    """Guarda las mejores soluciones en una nueva hoja del Excel"""
    
    print("\nğŸ’¾ Guardando soluciones en Excel...")
    
    balancer.load_workbook()
    
    # Crear nueva hoja "OptimizationResults" si no existe
    if "OptimizationResults" in balancer.wb.sheetnames:
        del balancer.wb["OptimizationResults"]
    
    results_sheet = balancer.wb.create_sheet("OptimizationResults")

    # ---- TÃTULO ----
    results_sheet.cell(1, 1, "Optimization Results")

    # ---- NOMBRES DE COLUMNAS (sin Rank) ----
    headers = [
        "Distance",
        "Player HP L12", "Player HP L3", "Player Dmg", "Player FireRate",
        "E1 HP", "E1 Spawn", "E1 FireRate",
        "E2 HP", "E2 Spawn",
        "E3 HP", "E3 Spawn", "E3 FireRate",
        "Boss HP", "ExtraLife Int", "Nuke Int",
        "SD Difficulty", "Survival Gap", "SD TTK", "Score Dev"
    ]
    
    for col, header in enumerate(headers, 1):
        results_sheet.cell(2, col, header)
    
    # ---- Calcular distancias y ordenar ----
    distances = np.sqrt(np.sum(objectives**2, axis=1))
    sorted_indices = np.argsort(distances)
    
    # ---- Escribir top 20 soluciones ----
    for rank, idx in enumerate(sorted_indices[:20], 1):
        row = rank + 2  # Empieza despuÃ©s de encabezados
        solution = solutions[idx]
        objective = objectives[idx]
        distance = distances[idx]
        
        # Columna 1: Distancia
        results_sheet.cell(row, 1, round(distance, 4))
        
        # ConfiguraciÃ³n (columnas 2â€“16)
        for col, value in enumerate(solution, 2):
            results_sheet.cell(row, col, round(value, 2))
        
        # Objetivos (columnas 17â€“20)
        for col, value in enumerate(objective, 17):
            results_sheet.cell(row, col, round(value, 4))
    
    # ---- Aplicar mejor soluciÃ³n ----
    balancer.write_config(solutions[best_idx])
    balancer.save_workbook()
    
    print("âœ… Soluciones guardadas en hoja 'OptimizationResults'")
    print("âœ… Mejor soluciÃ³n aplicada en hoja 'Config'")


def plot_pareto_front(pareto_front, best_idx):
    """Genera grÃ¡ficos del Pareto Front (como Figura 4 del paper)"""
    
    try:
        import matplotlib.pyplot as plt
        from mpl_toolkits.mplot3d import Axes3D
        
        print("\nğŸ“Š Generando grÃ¡ficos del Pareto Front...")
        
        fig = plt.figure(figsize=(15, 10))
        
        # GrÃ¡fico 1: SD Difficulty vs Survival Gap
        ax1 = fig.add_subplot(2, 2, 1)
        ax1.scatter(pareto_front[:, 0], pareto_front[:, 1], 
                   c='lightgreen', alpha=0.6, s=50, label='Soluciones Pareto')
        ax1.scatter(pareto_front[best_idx, 0], pareto_front[best_idx, 1],
                   c='red', marker='â˜…', s=300, label='Mejor SoluciÃ³n', zorder=5)
        ax1.set_xlabel('SD Difficulty (OBJ1)')
        ax1.set_ylabel('Survival Gap (OBJ2)')
        ax1.set_title('Pareto Front: Dificultad vs Supervivencia')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # GrÃ¡fico 2: SD TTK vs Score Deviation
        ax2 = fig.add_subplot(2, 2, 2)
        ax2.scatter(pareto_front[:, 2], pareto_front[:, 3],
                   c='lightgreen', alpha=0.6, s=50, label='Soluciones Pareto')
        ax2.scatter(pareto_front[best_idx, 2], pareto_front[best_idx, 3],
                   c='red', marker='â˜…', s=300, label='Mejor SoluciÃ³n', zorder=5)
        ax2.set_xlabel('SD TTK (OBJ3)')
        ax2.set_ylabel('Score Deviation (OBJ4)')
        ax2.set_title('Pareto Front: TTK vs Score')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        
        # GrÃ¡fico 3: 3D - Primeros 3 objetivos
        ax3 = fig.add_subplot(2, 2, 3, projection='3d')
        ax3.scatter(pareto_front[:, 0], pareto_front[:, 1], pareto_front[:, 2],
                   c='lightgreen', alpha=0.6, s=50)
        ax3.scatter(pareto_front[best_idx, 0], pareto_front[best_idx, 1], pareto_front[best_idx, 2],
                   c='red', marker='*', s=300)
        ax3.set_xlabel('SD Difficulty')
        ax3.set_ylabel('Survival Gap')
        ax3.set_zlabel('SD TTK')
        ax3.set_title('Pareto Front 3D')
        
        # GrÃ¡fico 4: Distancia al origen
        ax4 = fig.add_subplot(2, 2, 4)
        distances = np.sqrt(np.sum(pareto_front**2, axis=1))
        sorted_distances = np.sort(distances)
        ax4.plot(sorted_distances, 'o-', color='lightgreen', alpha=0.6)
        ax4.axhline(y=distances[best_idx], color='red', linestyle='--', 
                   label=f'Mejor: {distances[best_idx]:.3f}')
        ax4.set_xlabel('SoluciÃ³n (ordenada)')
        ax4.set_ylabel('Distancia al Origen')
        ax4.set_title('Calidad de Soluciones (menor = mejor)')
        ax4.legend()
        ax4.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig('pareto_front.png', dpi=300, bbox_inches='tight')
        print("âœ… GrÃ¡ficos guardados en 'pareto_front.png'")
        
        # Mostrar grÃ¡ficos
        plt.show()
        
    except ImportError:
        print("âš ï¸ matplotlib no disponible, grÃ¡ficos omitidos")
        print("   Instalar con: pip install matplotlib")


def manual_tuning_mode():
    """Modo de ajuste manual interactivo"""
    
    print("\nğŸ® MODO DE AJUSTE MANUAL")
    print("="*70)
    
    balancer = GalacticSiegeBalancer('game_balance.xlsx')
    balancer.load_workbook()
    
    current_config = balancer.read_current_config()
    
    print("\nConfiguraciÃ³n actual:")
    for key, value in current_config.items():
        print(f"  {key}: {value}")
    
    print("\nÂ¿QuÃ© deseas ajustar?")
    print("1. Player Health L1-L2")
    print("2. Enemy1 Health")
    print("3. Enemy1 Spawn Interval")
    print("4. Ver objetivos actuales")
    print("5. Aplicar ajustes recomendados del paper")
    print("0. Salir")
    
    choice = input("\nSelecciona opciÃ³n: ")
    
    if choice == "1":
        new_value = int(input("Nuevo valor para Player Health L1-L2 (3-10): "))
        config_array = list(current_config.values())
        config_array[0] = new_value
        objectives = balancer.evaluate_config(np.array(config_array))
        balancer.print_config(np.array(config_array), objectives)
        
    elif choice == "4":
        config_array = np.array(list(current_config.values()))
        objectives = balancer.evaluate_config(config_array)
        balancer.print_config(config_array, objectives)
        
    elif choice == "5":
        print("\nğŸ“‹ Aplicando ajustes recomendados...")
        recommended = np.array([
            8, 15,    # player health
            1, 0.4,   # player damage, fire rate
            4, 6, 2.0,  # enemy1
            3, 7,     # enemy2
            5, 10, 3, # enemy3
            40, 20, 40  # boss, powerups
        ])
        objectives = balancer.evaluate_config(recommended)
        balancer.print_config(recommended, objectives)
        
        apply = input("\nÂ¿Aplicar estos valores? (s/n): ")
        if apply.lower() == 's':
            balancer.write_config(recommended)
            print(" ConfiguraciÃ³n aplicada")


def export_to_cpp():
    """Exporta la configuraciÃ³n optimizada a cÃ³digo C++"""
    
    print("\n EXPORTAR A CÃ“DIGO C++")
    print("="*70)
    
    balancer = GalacticSiegeBalancer('game_balance.xlsx')
    balancer.load_workbook()
    config = balancer.read_current_config()
    
    cpp_code = f"""
// ========================================
// CONFIGURACIÃ“N OPTIMIZADA DE BALANCE
// Generada automÃ¡ticamente por balance_optimizer.py
// Fecha: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
// ========================================

// En LuaBinding.hpp - Enemy1Factory:
enemy1.AddComponent<HealthComponent>({config['enemy1_health']});
enemy1.AddComponent<ScoreComponent>(100);

// En LuaBinding.hpp - Enemy2Factory:
enemy2.AddComponent<HealthComponent>({config['enemy2_health']});
enemy2.AddComponent<ScoreComponent>(50);

// En LuaBinding.hpp - Enemy3Factory:
enemy3.AddComponent<HealthComponent>({config['enemy3_health']});
enemy3.AddComponent<ScoreComponent>(250);

// En scene_01.lua - Player:
health = {{
    health = {config['player_health_l12']}
}}

// En scene_03.lua - Player:
health = {{
    health = {config['player_health_l3']}
}}

// En player.lua:
local shootInterval = {config['player_fire_rate']}

// En enemy1.lua:
local shootInterval = {config['enemy1_fire_rate']}

// En enemyFactory_enemy1.lua:
local spawnInterval = {config['enemy1_spawn_interval']}

// En enemyFactory_enemy2.lua:
local spawnInterval = {config['enemy2_spawn_interval']}

// En enemy3.lua:
local shootInterval = {config['enemy3_fire_rate']}

// En enemyFactory_enemy3.lua:
local spawnInterval = {config['enemy3_spawn_interval']}

// En extraLifeFactory.lua:
local spawnInterval = {config['extra_life_interval']}

// En nukeFactory.lua:
local spawnInterval = {config['nuke_interval']}

// En scene_03.lua - Boss:
health = {{
    health = {config['boss_health']}
}}
"""
    
    # Guardar en archivo
    with open('optimized_config.cpp', 'w') as f:
        f.write(cpp_code)
    
    print(cpp_code)
    print("\n ConfiguraciÃ³n exportada a 'optimized_config.cpp'")


# ============================================
# MENÃš PRINCIPAL
# ============================================

def main():
    """MenÃº principal del optimizador"""
    
    print("""
â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
â•‘     GALACTIC SIEGE - BALANCE OPTIMIZER                       â•‘
â•‘     Basado en NSGA-II y el paper IEEE ICEBE 2021           â•‘
â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    """)
    
    while True:
        print("\n MENÃš PRINCIPAL:")
        print("1.  Ejecutar optimizaciÃ³n completa (NSGA-II)")
        print("2.  Modo ajuste manual")
        print("3.  Ver configuraciÃ³n actual")
        print("0.  Salir")
        
        choice = input("\nSelecciona opciÃ³n: ")
        
        if choice == "1":
            run_optimization()
        elif choice == "2":
            manual_tuning_mode()
        elif choice == "3":
            balancer = GalacticSiegeBalancer('game_balance.xlsx')
            balancer.load_workbook()
            config_array = np.array(list(balancer.read_current_config().values()))
            objectives = balancer.evaluate_config(config_array)
            balancer.print_config(config_array, objectives)
        else:
            print(" OpciÃ³n invÃ¡lida")


if __name__ == "__main__":
    
    main()
```
